```python
def temprature_converter(temp_value,unit):
    if unit == 'C':
        temprature =round((temp_value * 9/5) + 32, 2)
        print(f'{temp_value}°C is equal to {temprature}°F')
    elif unit == 'F':
        temprature = round((temp_value - 32) * 5/9, 2)
        print(f'{temp_value}°F is equal to {temprature}°C')
    else:
        print("Invalid unit. Please enter 'C' for Celsius or 'F' for Fahrenheit.")
```

```python
temprature_converter(25, 'C')
```

```python
def password_checker(password):
    if len(password) < 8:
        print("Password must be at least 8 characters long.")
        return False
    if password.islower() and password.isupper() and password.isdigit():
        print("Password is strong.")
        return True
    print("Password is not strong enough.")
    return False
```

```python
def password_checker(password):
    if len(password) < 8:
        print("Password must be at least 8 characters long.")
        return False
    if not any(char.isupper() for char in password):    
        print("Password must contain at least one uppercase letter.")
        return False
    if not any(char.islower() for char in password):    
        print("Password must contain at least one lowercase letter.")
        return False
    if not any(char.isdigit() for char in password):    
        print("Password must contain at least one digit.")
        return False
    print("Password is strong.")
    return True 
```

```python
password_checker("Abc12345")
```

```python
def palindrome_checker(word):
    cleaned_word = word.replace(" ", "").lower()
    if cleaned_word == cleaned_word[::-1]:
        print (f'thr word{word} is a palindrome.')
    else:
        print(f'thr word{word} is not a palindrome.') 
```

```python
def palindrome(string):
    string_values = string.replace(" ", "").lower()
    if string_values == string_values[::-1]:
        print(f'{string} is a palindrome.')
    else:
        print(f'{string} is not a palindrome.')
```

```python
cart = [
    {"name": "Laptop", "price": 999.99, "quantity": 1},
    {"name": "Headphones", "price": 199.99, "quantity": 2}, 
    {"name": "Mouse", "price": 49.99, "quantity": 1},

]
def cart_total(cart):
    total_cost = 0
    for item in cart:
        total_cost += item["price"] * item["quantity"]
    return round(total_cost, 2)
```

```python
print(f'Total cost of the cart is: ${cart_total(cart)}')
```
