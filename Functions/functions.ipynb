```python
def translator_return(english_text):
    if english_text == "hello":
        hindi_word = "Namaste"
    elif english_text == "Help":
        hindi_word = "madad"
    elif english_text == "How are you?":
        hindi_word = "aap kaise ho?"
    elif english_text == "Required":
        hindi_word = "jaruri"
    else:
        hindi_word = "Translation not found"
    
    print(f"the translation for {english_text} is {hindi_word}")
```

```python
english_return_input = input("Enter the english word you want to translate: ")
hindi_translation = translator_return(english_return_input)
print(f"The Hindi translation for '{english_return_input}' is: {hindi_translation}")
```

```python
def even_or_odd(num):
    if num % 2 == 0:
        print(f"{num} is an even number.")
    elif num % 2 != 0:
        print(f"{num} is an odd number.")
    elif type(num) != int:
        print("Please enter a valid number")
```

```python
final_output = even_or_odd(10)
print(final_output)
```

```python
def even_or_odd_instance(num):
    if not isinstance(num, int):
        return "Please enter a valid number"
    elif num % 2 == 0:
        return f"{num} is an even number."
    else:
        return f"{num} is an odd number."
```

```python
def factorial(num):
        if num == 0 or num == 1:
            print(f"the factorial of {num} is 1")
            return

        product = 1
        for i in range(1, num + 1):
            product *= i

        print(f"the factorial of {num} is {product}")
```

```python
number = int (input("Enter a number to find its factorial: "))
factorial(number)
```

```python
def factorial_recurssive(num):
    if num == 0 or num ==1:
        return 1
    else:
        return num* factorial_recurssive(num-1)
    
```

```python
factorial_recurssive(5)
```

```python
input_recursive = int(input("Enter a number to find its factorial using recursive function: "))
result_recursive = factorial_recurssive(input_recursive)
```
