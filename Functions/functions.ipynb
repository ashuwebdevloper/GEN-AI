```python
## Creating a function how translation works from english to hindilanguage'
def translator(english_text):
    if english_text == "hello":
        print(f"the translation for {english_text} is Namaste")
    if english_text=="Help":
        print(f"the translation for {english_text} is madad")
    if english_text=="How are you?":
        print(f"the translation for {english_text} is aap kaise ho?")
    if english_text=="Required":
        print(f"the translation for {english_text} is jaruri"



```

```python
english_input = input("Enter the english word you want to translate: ")
translator(english_input)
```

```python
def translator_return(english_text):
    if english_text == "hello":
        hindi_word = "Namaste"
    elif english_text == "Help":
        hindi_word = "madad"
    elif english_text == "How are you?":
        hindi_word = "aap kaise ho?"
    elif english_text == "Required":
        hindi_word = "jaruri"
    else:
        hindi_word = "Translation not found"
    
    print(f"the translation for {english_text} is {hindi_word}")
```

```python
english_return_input = input("Enter the english word you want to translate: ")
hindi_translation = translator_return(english_return_input)
print(f"The Hindi translation for '{english_return_input}' is: {hindi_translation}")
```
