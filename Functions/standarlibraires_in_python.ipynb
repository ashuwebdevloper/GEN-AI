```python
import csv
```

```python
with open('new_example.txt',mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'City'])
    writer.writerow(['Alice', 30, 'New York'])
```

```python
with open('new_example.txt', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
import json
```

```python
data = {
    "name": "Alice","age": 30,"city": "New York"}
jason_string = json.dumps(data)
print(type(jason_string))
print(jason_string)
```
