**MAP**


## list(map(function,iterable))

```python
def square(list_of_numbers):
    empty_list = []
    for item in list_of_numbers:
         squared_item = item ** 2
         empty_list.append(squared_item)
    return empty_list
numbers = [1,2,3,4,5]
print(square(numbers))
```

```python
print(list(map(lambda x:x**2, [1,2,3,4,5])))
```

```python
numbers_1 = [1,2,3,4,5]
numbers_2 = [6,7,8,9,10]
print(list(map(lambda x,y:x+y, numbers_1, numbers_2)))
```

```python
print(list(map(int, ['1','2','3','4','5'])))
```
