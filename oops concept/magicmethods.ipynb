```python
class avenger:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def __str__(self):
        return f"{self.name} + {self.power}"

    def __add__(self, other):
        return avenger(f"{self.name} & {other.name}", f"{self.power} and {other.power}")


ironman = avenger("ironman", "intelligence")
thor = avenger("thor", "thunder")
team = ironman + thor
print(team)
```
