### Polymorphism

```python
class Grandparent:
    def speak(self):
        print("Hello my name is amit")

    
class Parent(Grandparent):

    def speak(self):
        print("hello mh name is junior amit")

class child(Grandparent):

    def speak(self):
        print("Hello my name is junior amir junior")
```

```python
def attribute_provider(object):
    print(object.speak())
    
```

```python
granda=Grandparent()
```

```python
attribute_provider(granda)
```

```python
parent = Parent()
```

```python
attribute_provider(parent)
```

```python
child = child()
```

```python
attribute_provider(child)
```
