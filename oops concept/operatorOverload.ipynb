## Operator overloading concepts

works with use r defiend opbects
it makes custom object behaves like buillt in types

```python
class AIModel:
    def __init__(self,name,accuracy):
        self.name = name
        self.accuracy = accuracy
    def __gt__(self,other):
        return self.accuracy > other.accuracy

    def __lt__(self,other):
        return self.accuracy < other.accuracy

    def __eq__(self,other):
        return self.accuracy == other.accuracy
```

```python
gpt4=AIModel("GPT-4",0.95)
bard=AIModel("Bard",0.90)
```

```python
print(gpt4 > bard)  # True
```

```python
print(gpt4 < bard)  # True
```

```python
print(gpt4 == bard)  # True
```
