```python
class Avengerss:

    origin = "Avenger"

    def __init__(self,name,power,weapon,national):
        self.name = name
        self.power = power
        self.weapon = weapon    
        self.national=national

    @classmethod
    def show_origin(cls):
        print(f"All Avengers belong to {cls.origin} ")

    def shelf_info(self):
        print(f"my name is {self.name} and my power is {self.power} and my weapon is {self.weapon} and i am from {self.national}  ")
    

```

```python
iron_man = Avengerss("Ironman","Intelligence","Armour Suit","American")
```
