```python
class Grandparent:
    def __init__(self,nationality,baldness,anger_issues):
        self.nationality = nationality
        self.baldness = baldness
        self.anger_issues = anger_issues

    def introduction(self):
        print(f"I am from {self.nationality} and I have {self.baldness} and {self.anger_issues}.")
```

```python
class parent(Grandparent):
    def __init__(self,nationality,baldness,anger_issues,intelligence):
        super().__init__(nationality,baldness,anger_issues)
        self.intelligence=intelligence
    
    def intelligence_details(self):
        print(f"I am from {self.nationality} and I have {self.baldness} and {self.anger_issues}. My intelligence is {self.intelligence}.")
```

```python
class child(parent):
    def __init__(self, nationality, baldness, anger_issues, intelligence, curiosity):
        super().__init__(nationality, baldness, anger_issues, intelligence)
        self.curiosity = curiosity

    def curiosity_details(self):
        print(f"I am from {self.nationality} and I have {self.baldness} and {self.anger_issues}. My intelligence is {self.intelligence} and my curiosity is {self.curiosity}.")
```

```python
parent1 = parent("Indian", "Bald", "Mild", "High")
```

```python
parent1.introduction()
```

```python
child1 = child("Indian", "Bald", "low", "High", "Very High")
```

```python
child1.introduction()
```

```python
child1.intelligence_details()
```

```python
child1.curiosity_details()
```
