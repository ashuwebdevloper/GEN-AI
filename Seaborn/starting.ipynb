## Facet Plot (Grid Plots)
A Facet Plot allows us to split a single visualization into a grid of multiple subplots based on the unique categories of a categorical column.
Why use Faceting instead of hue?
When a categorical column has too many unique values, using hue overlays too many colors on a single graph, making it cluttered and unreadable.

Faceting solves this by giving each category its own dedicated, clean subplot space.

Key Mechanics:
Row-wise or Column-wise Separation: We can split the information across rows, columns, or both using the row and col parameters.

Figure-level Exclusive: As noted, faceting functionality is only possible using Figure-level functions (sns.relplot(), sns.displot(), sns.catplot()). It does not work with Axes-level functions (like sns.scatterplot()).

write about facet plot

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex', style='day')
```

## Distribution Plots
Distribution plots provide an overview of how data is distributed within a given dataset—typically targeting a Numerical Column.
Why use Distribution Plots?
Univariate Analysis: Ideal for looking at a single variable in isolation to understand its natural behavior.

Outlier Detection: Helps visually isolate extreme values (outliers) that lie far from the rest of the data.

Shape and Spread Analysis: Gives instant insight into whether the data follows a Normal Distribution (bell curve) or is Skewed (left/right tail), alongside revealing its overall spread and central tendency.

### 1. Histplot (sns.histplot)
A histogram checks the distribution of a numerical column by splitting the data into continuous, equal-interval intervals called bins. The height of each bar reflects the data count (frequency) within that specific interval.
Function Type:
Axes-Level Function: sns.histplot() (Draws directly onto an individual matplotlib axis)

Syntax and Key Parameters:
Python
sns.histplot(data=df, x="column_name", bins=n, kde=True)
data: The Pandas DataFrame containing your data.

x: The specific numerical column_name you want to analyze.

bins: Controls the resolution of your plot. You pass an integer n to specify the exact number of intervals you want inside the histogram.

Too few bins over-summarizes the data (hiding patterns).

Too many bins makes the plot jagged and noisy.

kde: A boolean flag (True/False). When set to True, it overlays a smooth Kernel Density Estimate (KDE) curve right on top of the histogram bars to show the overall continuous trend.
