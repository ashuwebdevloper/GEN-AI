### Machine Learning Pipeline WorkflowStep 1: Data CategorizationAction: Separating data into multiple categories.Step 2: Pipeline Transformer CreationAction: Create separate pipeline transformers tailored for each specific data category.Step 3: Apply TransformersAction: Apply these separate transformers directly to the corresponding separated data.Step 4: Model Creation PipelineAction: Construct the final model creation pipeline.Structure:$$\text{Pipeline} \longrightarrow \left[ \text{Features} + \text{Transformation} \right] \longrightarrow (\text{Algorithm})$$Step 5: Data Split & EncodingAction: Perform train_test_split on the data.Action: Encode the output/target column.Step 6: Training & EvaluationAction: Training the model and making predictions.

```python
Numeric_Data = ['Age','Balance','Day','Duration','Campagin','Pdays','Previous']

Ordinal_Data = ['Education']

Nominal_Data = ['Job','Marital','Default','Housing','Loan','contact','Month','poutcome']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

```python
Numeric_Pipeline = Pipeline(steps = [('impute',SimpleImputer(strategy='mean'))])
```

```python
Ordinal_Pipeline = Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),('encoding',OrdinalEncoder(categories=[['unknown','primary','secondary','tertiary']])),('scaling',StandardScaler())])
```

```python
Nominal_Pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),
    ('encoding', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ColumnTransformer(transformers=[
    ('Numeric Transformation', Numeric_Pipeline, Numeric_Data),
    ('Ordinal Transformation', Ordinal_Pipeline, Ordinal_Data),
    ('Nominal Transformation', Nominal_Pipeline, Nominal_Data),
])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('Numeric Transformation', Numeric_Pipeline, Numeric_Data),
    ('Ordinal Transformation', Ordinal_Pipeline, Ordinal_Data),
    ('Nominal Transformation', Nominal_Pipeline, Nominal_Data),
])

model = Pipeline(steps=[
    ('features', preprocessor),
    ('classifier', LogisticRegression())
])
```

##SVM Hyperparameter: C (Low Value / Soft Margin)
When the regularization parameter C is very small (less), the model allows for a softer margin.

Key Characteristics:
Focus of the Model: It focuses on finding a general classification boundary for the data points, rather than trying to perfectly separate every single point.

Handling Misclassifications:

It does not attempt to classify every point perfectly.

It accepts that some data points might still be misclassified or fall inside the margin.

Example Scenario:

Out of 1,000 data points, a few (e.g., 2 to 3 points) are allowed to be not properly classified.

Achieving an accuracy of around 97% or 98% is perfectly acceptable if it prevents overfitting.

Impact on Overfitting:
A smaller C acts as a stronger regularizer, widening the margin and reducing the risk of overfitting by ignoring minor noise or outliers.
