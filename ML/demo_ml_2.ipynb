```python
X = st[['Gender', 'Age', 'City', 'Profession', 'Academic Pressure', 'CGPA',
       'Study Satisfaction', 'Sleep Duration', 'Dietary Habits',
       'Work/Study Hours', 'Financial Stress',
       'Family History of Mental Illness',]]
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
```

```python
X = s.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Logistic Rigression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
y_predict = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,y_predict)
```

## Decision Tree Classifieer

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(max_depth=5)
```

```python
tree.fit(X_train,Y_train)
```

```python
y_predict=tree.predict(X_test)
```

```python
accuracy_score(Y_test,y_predict)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, Y_train)

rf_y_predict = rf.predict(X_test)
accuracy_score(Y_test, rf_y_predict)
```

```python
from catboost import CatBoostClassifier

cat = CatBoostClassifier(iterations=1000, learning_rate=0.1, depth=6,
                         random_state=42, verbose=0)
cat.fit(X_train, Y_train)

cat_y_predict = cat.predict(X_test)
print("CatBoost accuracy:", accuracy_score(Y_test, cat_y_predict))
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train, Y_train)

xgb_y_predict = xgb.predict(X_test)
print("XGBoost accuracy:", accuracy_score(Y_test, xgb_y_predict))
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
ada.fit(X_train, Y_train)

ada_y_predict = ada.predict(X_test)
print("AdaBoost accuracy:", accuracy_score(Y_test, ada_y_predict))
```

## KNeighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)

knn_y_predict = knn.predict(X_test)
print("KNN accuracy:", accuracy_score(Y_test, knn_y_predict))
```

## suppot vector classification

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_train, Y_train)

svm_y_predict = svm.predict(X_test)
print("SVC accuracy:", accuracy_score(Y_test, svm_y_predict))
```

## Naive Bayes 

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, Y_train)

nb_y_predict = nb.predict(X_test)
print("Naive Bayes accuracy:", accuracy_score(Y_test, nb_y_predict))
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
X_train_mm = mm.fit_transform(X_train)
X_test_mm = mm.transform(X_test)

mnb = MultinomialNB()
mnb.fit(X_train_mm, Y_train)

mnb_y_predict = mnb.predict(X_test_mm)
print("MultinomialNB accuracy:", accuracy_score(Y_test, mnb_y_predict))
```

## Bagging classifier

```python
from sklearn.ensemble import BaggingClassifier

bag = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42)
bag.fit(X_train, Y_train)

bag_y_predict = bag.predict(X_test)
print("Bagging Classifier accuracy:", accuracy_score(Y_test, bag_y_predict))
```

## Stacking classifier

```python
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier

estimators = [('rf', RandomForestClassifier()),('bc', BaggingClassifier())]



stack = StackingClassifier(estimators=estimators, final_estimator=AdaBoostClassifier())
stack.fit(X_train, Y_train)

stack_y_predict = stack.predict(X_test)
print("Stacking accuracy:", accuracy_score(Y_test, stack_y_predict))
```
